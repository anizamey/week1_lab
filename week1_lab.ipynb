```python
w = [0.2 , -0.1]
b = 0.0
x = [1.0 , 2.0]
y_true = 1
lr = 0.5

#yˆ = { 1 ifw · x + b ≥ 00otherwise
#w = w + lr*(y true - y hat)*x, b = b + lr*(y true -
#y hat)

z = sum(wi*xi for wi, xi in zip(w,x)) + b

if z >=0:
    y_pred=1
else:
    y_pred=0
print('Prediction:',y_pred)

#if y_pred != y_true:
error = y_true - y_pred
w= [wi + lr * error * xi for wi, xi in zip(w,x)]
b += lr * error
print(f"Updated w: {w}")
print(f"Updated b: {b}")
```

Problem 9

```python
X = [
[1 , 1] ,
[2 , 1] ,
[2 , 3] ,
[3 , 3]
]
y = [0 , 0 , 1 , 1]

w=[0,0]
b=0
lr=0.1
epochs=10

for epoch in range(epochs):
  for i in range(len(X)):
    #\(z=w_{1}x_{1}+w_{2}x_{2}+b\)
    z = w[0] * X[i][0] + X[i][1] * w[1] + b
    # step to check and correcting
    if z >=0:
      y_pred=1
    else:
      y_pred=0

    if y_pred != y[i]:
      error = y[i] - y_pred
      w[0] += lr * error * X[i][0]
      w[1] += lr * error * X[i][1]
      b += lr * error
# step 3
correct = 0
for i in range(len(X)):
    z = w[0] * X[i][0] + w[1] * X[i][1] + b
    y_pred = 1 if z >= 0 else 0
    if y_pred == y[i]:
        correct += 1

accuracy = correct / len(X)

print(f"Final w: {w}")
print(f"Final b: {b:.1f}")
print(f"Accuracy: {accuracy}")
```
